# Vissualisation of the results

Import libraries.

```python
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
from pathlib import Path
```

Define utils functions.

```python
def get_slice(case: int, path: str) -> np.array:
    data = nib.load((path / 'Case_{}.nii.gz'.format(case)).resolve()).get_fdata()
    return data[:, :, int(data.shape[2]/2)]

def display(path):
    fig, ax = plt.subplots(10, 5, figsize=(100, 100))
    ax = ax.ravel()
    for i, case in enumerate(range(101, 151)):
        ax[i].set_title("Case {} ".format(case), fontsize=60)
        ax[i].imshow(get_slice(case, path))
```

## Uncertainty-based submission

Here are the middle slices of every case in the test set for the uncertainty-based submission.

```python
path = Path("uncertainty_based_submission") / "prediction"
display(path)
```
